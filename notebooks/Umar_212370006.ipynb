```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('data.csv')
x = data['Country Name ']
y = data['GDP per Capita ($)']

c = ['r', 'b', 'g', 'y', 'r']

plt.xlabel('Country Data')
plt.ylabel('GDP per Capita')
plt.title('GDP per Capita $ by Country Data ')

plt.bar(x, y, color=c, alpha=0.5, linewidth=4, width=0.5, align='center')
```
